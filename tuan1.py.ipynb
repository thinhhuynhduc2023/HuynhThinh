```python
d = float(input("Nhap duong kinh (mm): "))
f = float(input("Nhap luc (N): "))
if d <= 0 or f <= 0:
    print("Du lieu khong hop le")
else:
    s = d / 2 * d / 2 * 3.14159
    ung_suat = f / s
    print("Ung suat: ", ung_suat, "(MPa)")
    if ung_suat < 100:
        print(" AN TOAN - he so du tru cao")
    elif 100 <= ung_suat < 200:
        print("Chap nhan duoc, trong gioi han cho phep")
    else:
        print("KHONG DAT - vuot ung suat cho phep - can tang duong kinh")
```

```python
def ham_kiem_tra(d, f):
   if d > 0 and f > 0:
       return True
   else:
       return False
def dien_tich(d):
    return d / 2 * d / 2 * 3.14159
def ung_suat(s,f):
    return f / s
def phan_loai(ung_suat, gioi_han=200):
    if ung_suat < gioi_han / 2:
        return "An toan"
    elif ung_suat < gioi_han:
        return "Chap nhan duoc"
    else:
        return "Khong dat"
d = float(input("Nhap gia tri duong kinh (mm)"))
f = float(input("Nhap gia tri luc (N)"))
if  ham_kiem_tra(d,f):
    s = dien_tich(d)
    us = ung_suat(s,f)
    for d_test in range(10, 51, 5):
        s = d_test / 2 * d_test / 2 * 3.14159
        us = ung_suat(s,f)
        ket_qua_1 = phan_loai(us)
        ket_qua_150 = phan_loai(us, gioi_han=150)
        print ("D =", d_test, "-> Dien tich = ", s)
        print ("Ung suat = ", us, "MPa")
        print("Gioi han 200: ", ket_qua_1)
        print("Gioi han 150: ", ket_qua_150)
        print("---------------------------------------------------")
else:
    print("Duong kinh va luc phai lon hon 0, hay nhap lai so lieu")
```

```python
class ChiTietTru:
    def __init__(self, duong_kinh, luc):
        self.duong_kinh = duong_kinh
        self.luc = luc
    def tinh_dien_tich(self):
        return self.duong_kinh / 2 * self.duong_kinh / 2 * 3.14159
    def tinh_ung_suat(self):
        return self.luc / self.tinh_dien_tich()
    def phan_loai(self, gioi_han=200):
        if self.tinh_ung_suat() < gioi_han / 2:
            return "AN TOAN"
        elif self.tinh_ung_suat() < gioi_han:
            return "CHAP NHAN DUOC"
        else:
            return "KHONG DAT"
truc1 = ChiTietTru(25, 100000)
truc2 = ChiTietTru(40, 80000)
truc3 = ChiTietTru(30, 110000)
for truc in [truc1, truc2, truc3]:
    print("D = ", truc.duong_kinh, "Ung suat = ", truc.tinh_ung_suat(), "MPa ->", truc.phan_loai())
```

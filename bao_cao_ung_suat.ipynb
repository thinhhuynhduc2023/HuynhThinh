```python
class ChiTietTru:
    def __init__(self, ten, duong_kinh, luc):
        self.ten = ten
        self.duong_kinh = duong_kinh
        self.luc = luc

    def tinh_dien_tich(self):
        return self.duong_kinh / 2 * self.duong_kinh / 2 * 3.14159

    def tinh_ung_suat(self):
        return self.luc / self.tinh_dien_tich()

    def phan_loai(self, gioi_han=200):
        us = self.tinh_ung_suat()
        if us < gioi_han / 2:
            return "AN TOAN"
        elif us < gioi_han:
            return "CHAP NHAN DUOC"
        else:
            return "KHONG DAT"

danh_sach_truc = [
    ChiTietTru("Truc A", 25, 100000),
    ChiTietTru("Truc B", 40, 80000),
    ChiTietTru("Truc C", 30, 150000),
]

with open("bao_cao2.txt", "w") as f:
    f.write("BAO CAO TINH UNG SUAT\n")
    f.write("--------------------\n")
    for truc in danh_sach_truc:
        us = truc.tinh_ung_suat()
        dong = f"{truc.ten}: D={truc.duong_kinh}mm, Ung suat={us:.2f}MPa -> {truc.phan_loai()}\n"
        f.write(dong)
        print(dong, end="")   # vừa ghi file vừa hiện lên màn hình để kiểm tra
```
